# Q. Statistical Mechanics of the Quantum Oscillator

Consider a single quantum harmonic oscillator of angular frequency $\omega$ in thermal equilibrium with a heat bath at temperature $T$. The quantum mechanical energy eigenvalues are given by:

$$E_j = \hbar \omega \left(j + \frac{1}{2}\right), \quad \text{for } j = 0, 1, 2, \dots$$

where $\hbar$ is the reduced Planck constant and $k_B$ is the Boltzmann constant.

**Answer the Following:**

1. **The Partition Function:**
Show that the canonical partition function $Z(T) = \sum_{j=0}^\infty e^{-E_j / (k_B T)}$ forms a convergent geometric series that simplifies to:

$$Z(T) = \frac{e^{-\frac{\hbar \omega}{2k_BT}}}{1 - e^{-\frac{\hbar \omega}{k_BT}}}$$


2. **Thermodynamic Potentials:**
Using your expression for $Z(T)$, derive explicit formulas for:
* The **Helmholtz Free Energy**, $F(T) = -k_B T \ln Z$.
* The **Internal Energy**, $U(T) = -\frac{\partial \ln Z}{\partial \beta}$ (where $\beta = \frac{1}{k_B T}$).


3. **Entropy:**
Find the **Entropy** $S(T)$ of the system using the thermodynamic relation $S = -\left(\frac{\partial F}{\partial T}\right)$, and verify that your result satisfies $U = F + TS$.
4. **Heat Capacity and Hyperbolic Form:**
* Calculate the constant-volume **Heat Capacity**, $C_V = \left(\frac{\partial U}{\partial T}\right)$.
* Use the hyperbolic identity $\sinh(x) = \frac{e^x - e^{-x}}{2}$ to prove that your expression for $C_V$ can be compactly rewritten as:

$$C_V = k_B \left(\frac{\hbar \omega}{2k_BT}\right)^2 \frac{1}{\sinh^2\left(\frac{\hbar \omega}{2k_BT}\right)}$$




5. **Asymptotic Behavior (Discussion):**
Briefly describe the behavior of the heat capacity $C_V$ in the high-temperature limit ($k_B T \gg \hbar \omega$). Does it match the classical equipartition theorem prediction?

 Q. Statistical Mechanics of the Quantum Oscillator

## 1. The Partition Function

The energy eigenvalues of a quantum harmonic oscillator are

$$
E_j=\hbar\omega\left(j+\frac12\right), \qquad j=0,1,2,\ldots
$$

The canonical partition function is

$$
Z(T)=\sum_{j=0}^{\infty}e^{-E_j/(k_BT)}.
$$

Substituting the energy eigenvalues,

$$
Z
=
\sum_{j=0}^{\infty}
\exp\!\left[
-\frac{\hbar\omega}{k_BT}
\left(j+\frac12\right)
\right].
$$

Factor out the constant term:

$$
Z
=
e^{-\frac{\hbar\omega}{2k_BT}}
\sum_{j=0}^{\infty}
\left(
e^{-\frac{\hbar\omega}{k_BT}}
\right)^j.
$$

This is a geometric series with common ratio

$$
r=e^{-\frac{\hbar\omega}{k_BT}},
$$

where

$$
0<r<1,
$$

so the series converges:

$$
\sum_{j=0}^{\infty}r^j=\frac{1}{1-r}.
$$

Therefore,

$$
\boxed{
Z(T)
=
\frac{
e^{-\frac{\hbar\omega}{2k_BT}}
}{
1-e^{-\frac{\hbar\omega}{k_BT}}
}
}
$$

---

## 2. Thermodynamic Potentials

Let

$$
\beta=\frac{1}{k_BT}.
$$

Then,

$$
Z
=
\frac{
e^{-\beta\hbar\omega/2}
}{
1-e^{-\beta\hbar\omega}
}.
$$

Taking the natural logarithm,

$$
\ln Z
=
-\frac{\beta\hbar\omega}{2}
-
\ln\left(1-e^{-\beta\hbar\omega}\right).
$$

### (a) Helmholtz Free Energy

Using

$$
F=-k_BT\ln Z,
$$

we obtain

$$
F
=
-k_BT
\left[
-\frac{\hbar\omega}{2k_BT}
-
\ln\left(
1-e^{-\frac{\hbar\omega}{k_BT}}
\right)
\right].
$$

Hence,

$$
\boxed{
F
=
\frac{\hbar\omega}{2}
+
k_BT
\ln\left(
1-e^{-\frac{\hbar\omega}{k_BT}}
\right)
}
$$

### (b) Internal Energy

Using

$$
U
=
-\frac{\partial\ln Z}{\partial\beta},
$$

differentiate:

$$
\frac{\partial\ln Z}{\partial\beta}
=
-\frac{\hbar\omega}{2}
-
\frac{\hbar\omega e^{-\beta\hbar\omega}}
{1-e^{-\beta\hbar\omega}}.
$$

Therefore,

$$
U
=
\frac{\hbar\omega}{2}
+
\frac{\hbar\omega}
{e^{\beta\hbar\omega}-1}.
$$

Replacing

$$
\beta=\frac1{k_BT},
$$

gives

$$
\boxed{
U
=
\frac{\hbar\omega}{2}
+
\frac{\hbar\omega}
{e^{\frac{\hbar\omega}{k_BT}}-1}
}
$$

---

## 3. Entropy

The entropy is

$$
S
=
-
\left(
\frac{\partial F}{\partial T}
\right).
$$

Starting from

$$
F
=
\frac{\hbar\omega}{2}
+
k_BT
\ln
\left(
1-e^{-\frac{\hbar\omega}{k_BT}}
\right),
$$

differentiate with respect to \(T\):

$$
S
=
-k_B
\ln
\left(
1-e^{-\frac{\hbar\omega}{k_BT}}
\right)
+
\frac{\hbar\omega}{T}
\frac{1}
{e^{\frac{\hbar\omega}{k_BT}}-1}.
$$

Thus,

$$
\boxed{
S
=
-k_B
\ln
\left(
1-e^{-\frac{\hbar\omega}{k_BT}}
\right)
+
\frac{\hbar\omega}{T}
\frac{1}
{e^{\frac{\hbar\omega}{k_BT}}-1}
}
$$

### Verification that \(U=F+TS\)

Substitute the expressions for \(F\) and \(S\):

$$
\begin{aligned}
F+TS
&=
\frac{\hbar\omega}{2}
+
k_BT
\ln\left(1-e^{-\frac{\hbar\omega}{k_BT}}\right) \\
&\quad
+
T
\left[
-k_B
\ln\left(1-e^{-\frac{\hbar\omega}{k_BT}}\right)
+
\frac{\hbar\omega}{T}
\frac1{e^{\frac{\hbar\omega}{k_BT}}-1}
\right] \\
&=
\frac{\hbar\omega}{2}
+
\frac{\hbar\omega}
{e^{\frac{\hbar\omega}{k_BT}}-1} \\
&=U.
\end{aligned}
$$

Hence,

$$
\boxed{U=F+TS.}
$$

---

## 4. Heat Capacity

The constant-volume heat capacity is

$$
C_V
=
\left(
\frac{\partial U}{\partial T}
\right).
$$

Let

$$
x
=
\frac{\hbar\omega}{k_BT}.
$$

Then

$$
U
=
\frac{\hbar\omega}{2}
+
\frac{\hbar\omega}{e^x-1}.
$$

Since

$$
\frac{dx}{dT}
=
-\frac{x}{T},
$$

we obtain

$$
\boxed{
C_V
=
k_B
\left(
\frac{\hbar\omega}{k_BT}
\right)^2
\frac{e^x}
{(e^x-1)^2}
}
$$

where

$$
x=\frac{\hbar\omega}{k_BT}.
$$

### Hyperbolic Form

Using the identity

$$
\sinh(y)
=
\frac{e^y-e^{-y}}2,
$$

let

$$
y=\frac{x}{2}.
$$

Then,

$$
\sinh^2(y)
=
\frac{(e^x-1)^2}{4e^x},
$$

which gives

$$
\frac{e^x}{(e^x-1)^2}
=
\frac1{4\sinh^2(x/2)}.
$$

Substituting this into the previous expression,

$$
\boxed{
C_V
=
k_B
\left(
\frac{\hbar\omega}{2k_BT}
\right)^2
\frac1{\sinh^2\left(
\frac{\hbar\omega}{2k_BT}
\right)}
}
$$

---

## 5. Asymptotic Behavior (Discussion)

In the high-temperature limit,

$$
k_BT\gg\hbar\omega,
$$

or equivalently,

$$
x=\frac{\hbar\omega}{k_BT}\ll1.
$$

Using the approximation

$$
\sinh\left(\frac{x}{2}\right)
\approx
\frac{x}{2},
$$

the heat capacity becomes

$$
C_V
\approx
k_B
\left(\frac{x}{2}\right)^2
\frac1{(x/2)^2}
=
k_B.
$$

Therefore,

$$
\boxed{
\lim_{T\to\infty}C_V=k_B.
}
$$

This agrees exactly with the **classical equipartition theorem**, which predicts that a one-dimensional harmonic oscillator has an average thermal energy of

$$
U=k_BT,
$$

so its heat capacity is

$$
\boxed{C_V=k_B.}
$$

# Q. The Maximum Entropy Distribution under Mean and Variance Constraints

Consider the random variable $X \in (-\infty, \infty)$. Let:

1. **The Mean:** $\mathbb{E}[X] = \mu$
2. **The Variance:** $\mathbb{E}[(X-\mu)^2] = \sigma^2$

Show that the law of $X$ that maximizes entropy is given by:

$$X \sim \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

## ** Answers - The Maximum Entropy Distribution under Mean and Variance Constraints**

To find the probability density function $f(x)$ that maximizes the differential entropy, we set up a constrained optimization problem using **Lagrange Multipliers**.

### **1. Objective Function and Constraints**

The differential entropy of a continuous random variable $X \in (-\infty, \infty)$ is defined as:
$$H(f) = -\int_{-\infty}^{\infty} f(x) \ln f(x) \, dx$$

We maximize $H(f)$ subject to three constraints:
1. **Total Probability:** $\int_{-\infty}^{\infty} f(x) \, dx = 1$
2. **Mean Constraint:** $\int_{-\infty}^{\infty} x f(x) \, dx = \mu$
3. **Variance Constraint:** $\int_{-\infty}^{\infty} (x - \mu)^2 f(x) \, dx = \sigma^2$

---

### **2. The Lagrangian Formulation**

We construct the Lagrangian function $L$ with multipliers $\lambda_0, \lambda_1,$ and $\lambda_2$:

$$L(f, \lambda_0, \lambda_1, \lambda_2) = -\int_{-\infty}^{\infty} f(x) \ln f(x) \, dx - \lambda_0 \left( \int_{-\infty}^{\infty} f(x) \, dx - 1 \right) - \lambda_1 \left( \int_{-\infty}^{\infty} x f(x) \, dx - \mu \right) - \lambda_2 \left( \int_{-\infty}^{\infty} (x - \mu)^2 f(x) \, dx - \sigma^2 \right)$$

Combining the integrals, we get:

$$L = \int_{-\infty}^{\infty} \left[ -f(x) \ln f(x) - \lambda_0 f(x) - \lambda_1 x f(x) - \lambda_2 (x - \mu)^2 f(x) \right] dx + \lambda_0 + \lambda_1 \mu + \lambda_2 \sigma^2$$

---

### **3. Optimizing Using Functional Calculus**

Taking the functional derivative of $L$ with respect to $f(x)$ and setting it to zero:

$$\frac{\partial}{\partial f(x)} \left[ -f(x) \ln f(x) - \lambda_0 f(x) - \lambda_1 x f(x) - \lambda_2 (x - \mu)^2 f(x) \right] = 0$$

$$- \ln f(x) - 1 - \lambda_0 - \lambda_1 x - \lambda_2 (x - \mu)^2 = 0$$

Solving for $\ln f(x)$:

$$\ln f(x) = -(1 + \lambda_0) - \lambda_1 x - \lambda_2 (x - \mu)^2$$

Exponentiating both sides gives the functional form of our distribution:

$$f(x) = \exp\left( -(1 + \lambda_0) - \lambda_1 x - \lambda_2 (x - \mu)^2 \right)$$

---

### **4. Solving for the Multipliers**

#### **Step 4.1: Eliminate $\lambda_1$ using the Mean**
Since the distribution must be symmetric around its mean $\mu$, the linear term involving $x$ outside of the squared deviation must vanish to satisfy $\mathbb{E}[X] = \mu$. Therefore:
$$\lambda_1 = 0$$

This simplifies our density function to:
$$f(x) = e^{-(1 + \lambda_0)} \exp\left( -\lambda_2 (x - \mu)^2 \right)$$

#### **Step 4.2: Normalization Constraint ($\lambda_0$)**
Substitute $f(x)$ into the total probability constraint:
$$\int_{-\infty}^{\infty} e^{-(1 + \lambda_0)} \exp\left( -\lambda_2 (x - \mu)^2 \right) dx = 1$$

Using the standard Gaussian integral identity $\int_{-\infty}^{\infty} e^{-a(x-b)^2} dx = \sqrt{\frac{\pi}{a}}$:
$$e^{-(1 + \lambda_0)} \sqrt{\frac{\pi}{\lambda_2}} = 1 \implies e^{-(1 + \lambda_0)} = \sqrt{\frac{\lambda_2}{\pi}}$$

Now we can rewrite $f(x)$ in terms of $\lambda_2$ only:
$$f(x) = \sqrt{\frac{\lambda_2}{\pi}} \exp\left( -\lambda_2 (x - \mu)^2 \right)$$

#### **Step 4.3: Variance Constraint ($\lambda_2$)**
Substitute the updated $f(x)$ into the variance constraint:
$$\int_{-\infty}^{\infty} (x - \mu)^2 \sqrt{\frac{\lambda_2}{\pi}} \exp\left( -\lambda_2 (x - \mu)^2 \right) dx = \sigma^2$$

Using the standard identity $\int_{-\infty}^{\infty} y^2 e^{-a y^2} dy = \frac{1}{2a}\sqrt{\frac{\pi}{a}}$ where $y = x - \mu$ and $a = \lambda_2$:
$$\sqrt{\frac{\lambda_2}{\pi}} \cdot \left( \frac{1}{2\lambda_2} \sqrt{\frac{\pi}{\lambda_2}} \right) = \sigma^2$$

$$\frac{1}{2\lambda_2} = \sigma^2 \implies \lambda_2 = \frac{1}{2\sigma^2}$$

---

### **5. Final Conclusion**

Substituting $\lambda_2 = \frac{1}{2\sigma^2}$ back into our expression for $f(x)$:

$$f(x) = \sqrt{\frac{1}{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

Thus, the law of $X$ that maximizes entropy under fixed mean and variance constraints is uniquely the **Normal (Gaussian) Distribution**:

$$X \sim \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

# Q. Multiclass Logistic Regression: Automatic Classification of Emails

An e-commerce platform automatically routes inbound customer support emails to one of four mutually exclusive target destinations ($Y$):

* $Y = 0$ : **Finance**
* $Y = 1$ : **Technical**

Before final classification occurs, an upstream **AI Feature Extractor Agent** processes the raw text of each incoming email and outputs a discrete binary state observation vector $\mathbf{x} = (x_1, x_2, x_3)^T \in \mathscr{X}$, where:

* $x_1 = \mathbb{I}(\text{Financial context detected})$ (e.g., words like *"charge"*, *"invoice"*)
* $x_2 = \mathbb{I}(\text{Account access context detected})$ (e.g., words like *"password"*, *"login"*)
* $x_3 = \mathbb{I}(\text{Bug/malfunction context detected})$ (e.g., words like *"error"*, *"frozen"*)

---

Because human language inherently overlaps, the AI Agent's outputs are distributed non-deterministically across three common operational scenarios. The marginal probability of observing each state is fixed by historical logs to:

1. **Scenario A (Pure Financial):** $\mathbf{x}_A = (1, 0, 0)^T \implies P(\mathbf{x}_A) = 0.40$
2. **Scenario B (Pure Acces):** $\mathbf{x}_B = (0, 1, 0)^T \implies P(\mathbf{x}_B) = 0.25$
3. **Scenario C (Pure Technical):** $\mathbf{x}_C = (0, 0, 1)^T \implies P(\mathbf{x}_C) = 0.05$
3. **Scenario D (Access Technical):** $\mathbf{x}_D = (0, 1, 1)^T \implies P(\mathbf{x}_D) = 0.15$
3. **Scenario E (Billing Access):** $\mathbf{x}_E = (1, 1, 0)^T \implies P(\mathbf{x}_E) = 0.05$

---

Historical manual audits provide the following observed joint dataset expectations:

* Constraint 1: $\mathbb{E}[x_1 \cdot \mathbb{I}(Y=1)] = \mathbf{0.05}$
(Very low. If financial words are present, it is rarely routed to Technical support.)

* Constraint 2: $\mathbb{E}[x_2 \cdot \mathbb{I}(Y=1)] = \mathbf{0.15}$
(Moderate. Account access issues split down the middle depending on context.)

* Constraint 3: $\mathbb{E}[x_3 \cdot \mathbb{I}(Y=1)] = \mathbf{0.16}$
(High. If a bug or malfunction context is detected, it goes to the Tech team the vast majority of the time, utilizing $80\%$ of the $0.20$ max capacity).




## ** Answers - Multiclass Logistic Regression: Automatic Classification of Emails**

We want to find the conditional probability distribution $P(Y=1 \mid \mathbf{x})$ that maximizes the conditional entropy subject to the given historical constraints. For a binary outcome $Y \in \{0,1\}$, the maximum entropy solution takes the form of a logistic regression model:

$$P(Y=1 \mid \mathbf{x}) = \sigma(\mathbf{w}^T \mathbf{x}) = \frac{1}{1 + \exp(-(\omega_1 x_1 + \omega_2 x_2 + \omega_3 x_3))}$$

Where $\mathbf{w} = (\omega_1, \omega_2, \omega_3)^T$ are the weights (Lagrange multipliers) corresponding to each feature constraint.

---

### **1. Operational Scenarios and Joint Probabilities**

The problem specifies 5 active scenarios for the feature vector $\mathbf{x} = (x_1, x_2, x_3)^T$. Let's write out the conditional probabilities $P(Y=1 \mid \mathbf{x})$ for each scenario using our model parameters:

* **Scenario A:** $\mathbf{x}_A = (1, 0, 0)^T \implies P(Y=1 \mid \mathbf{x}_A) = \frac{1}{1 + e^{-\omega_1}}$
* **Scenario B:** $\mathbf{x}_B = (0, 1, 0)^T \implies P(Y=1 \mid \mathbf{x}_B) = \frac{1}{1 + e^{-\omega_2}}$
* **Scenario C:** $\mathbf{x}_C = (0, 0, 1)^T \implies P(Y=1 \mid \mathbf{x}_C) = \frac{1}{1 + e^{-\omega_3}}$
* **Scenario D:** $\mathbf{x}_D = (0, 1, 1)^T \implies P(Y=1 \mid \mathbf{x}_D) = \frac{1}{1 + e^{-(\omega_2 + \omega_3)}}$
* **Scenario E:** $\mathbf{x}_E = (1, 1, 0)^T \implies P(Y=1 \mid \mathbf{x}_E) = \frac{1}{1 + e^{-(\omega_1 + \omega_2)}}$

---

### **2. Setting up the Constraint Equations**

The empirical expectations are given by $\mathbb{E}[x_i \cdot \mathbb{I}(Y=1)] = \sum_{\mathbf{x}} P(\mathbf{x}) \cdot x_i \cdot P(Y=1 \mid \mathbf{x})$.

#### **Constraint 1: Financial Context ($x_1$)**
This feature is active only in **Scenario A** and **Scenario E**.
$$\mathbb{E}[x_1 \cdot \mathbb{I}(Y=1)] = P(\mathbf{x}_A)P(Y=1 \mid \mathbf{x}_A) + P(\mathbf{x}_E)P(Y=1 \mid \mathbf{x}_E) = 0.05$$
$$0.40 \left( \frac{1}{1 + e^{-\omega_1}} \right) + 0.05 \left( \frac{1}{1 + e^{-(\omega_1 + \omega_2)}} \right) = 0.05 \quad \text{--- (Eq. 1)}$$

#### **Constraint 2: Account Access Context ($x_2$)**
This feature is active in **Scenario B**, **Scenario D**, and **Scenario E**.
$$\mathbb{E}[x_2 \cdot \mathbb{I}(Y=1)] = P(\mathbf{x}_B)P(Y=1 \mid \mathbf{x}_B) + P(\mathbf{x}_D)P(Y=1 \mid \mathbf{x}_D) + P(\mathbf{x}_E)P(Y=1 \mid \mathbf{x}_E) = 0.15$$
$$0.25 \left( \frac{1}{1 + e^{-\omega_2}} \right) + 0.15 \left( \frac{1}{1 + e^{-(\omega_2 + \omega_3)}} \right) + 0.05 \left( \frac{1}{1 + e^{-(\omega_1 + \omega_2)}} \right) = 0.15 \quad \text{--- (Eq. 2)}$$

#### **Constraint 3: Bug/Malfunction Context ($x_3$)**
This feature is active in **Scenario C** and **Scenario D**.
$$\mathbb{E}[x_3 \cdot \mathbb{I}(Y=1)] = P(\mathbf{x}_C)P(Y=1 \mid \mathbf{x}_C) + P(\mathbf{x}_D)P(Y=1 \mid \mathbf{x}_D) = 0.16$$
$$0.05 \left( \frac{1}{1 + e^{-\omega_3}} \right) + 0.15 \left( \frac{1}{1 + e^{-(\omega_2 + \omega_3)}} \right) = 0.16 \quad \text{--- (Eq. 3)}$$

---

### **3. Solving the System for $P(Y=1 \mid \mathbf{x})$**

Let's look closely at **Constraint 3 (Eq. 3)**. The maximum possible value for the first term is $0.05$ (if $P(Y=1|\mathbf{x}_C)=1$) and for the second term is $0.15$ (if $P(Y=1|\mathbf{x}_D)=1$), summing to a maximum capacity of $0.20$. We are told it utilizes $80\%$ of this capacity ($0.16$).

Let's test the natural operational boundary where a pure bug/malfunction ($\mathbf{x}_C$) or an access bug ($\mathbf{x}_D$) almost certainly routes to Technical support ($Y=1$):
$$\text{Assume } P(Y=1 \mid \mathbf{x}_C) = 1 \implies \omega_3 \to \infty$$

Substituting this into Eq. 3:
$$0.05(1) + 0.15 P(Y=1 \mid \mathbf{x}_D) = 0.16 \implies 0.15 P(Y=1 \mid \mathbf{x}_D) = 0.11 \implies P(Y=1 \mid \mathbf{x}_D) = \frac{11}{15} \approx 0.7333$$

Now, let's substitute $P(Y=1 \mid \mathbf{x}_D) = \frac{11}{15}$ into **Eq. 2**:
$$0.25 P(Y=1 \mid \mathbf{x}_B) + 0.15 \left(\frac{11}{15}\right) + 0.05 P(Y=1 \mid \mathbf{x}_E) = 0.15$$
$$0.25 P(Y=1 \mid \mathbf{x}_B) + 0.11 + 0.05 P(Y=1 \mid \mathbf{x}_E) = 0.15 \implies 0.25 P(Y=1 \mid \mathbf{x}_B) + 0.05 P(Y=1 \mid \mathbf{x}_E) = 0.04 \quad \text{--- (Eq. 4)}$$

Now let's check **Eq. 1**:
$$0.40 P(Y=1 \mid \mathbf{x}_A) + 0.05 P(Y=1 \mid \mathbf{x}_E) = 0.05 \quad \text{--- (Eq. 5)}$$

Because financial contexts rarely map to Technical support, let's find the solution where a pure financial email goes to Finance ($Y=0$), meaning $P(Y=1 \mid \mathbf{x}_A) = 0 \implies \omega_1 \to -\infty$.

Substituting $P(Y=1 \mid \mathbf{x}_A) = 0$ into Eq. 5:
$$0.05 P(Y=1 \mid \mathbf{x}_E) = 0.05 \implies P(Y=1 \mid \mathbf{x}_E) = 1$$

Now substitute $P(Y=1 \mid \mathbf{x}_E) = 1$ back into Eq. 4 to solve for Scenario B:
$$0.25 P(Y=1 \mid \mathbf{x}_B) + 0.05(1) = 0.04 \implies 0.25 P(Y=1 \mid \mathbf{x}_B) = -0.01$$

Since a probability cannot be negative, $P(Y=1 \mid \mathbf{x}_A)$ cannot be exactly $0$. Let's solve the exact algebraic system for the unique valid probability values:

From Eq. 5: $P(Y=1 \mid \mathbf{x}_E) = 1 - 8P(Y=1 \mid \mathbf{x}_A)$  
From Eq. 4: $0.25 P(Y=1 \mid \mathbf{x}_B) + 0.05(1 - 8P(Y=1 \mid \mathbf{x}_A)) = 0.04 \implies 0.25 P(Y=1 \mid \mathbf{x}_B) - 0.40P(Y=1 \mid \mathbf{x}_A) = -0.01$

Setting $P(Y=1 \mid \mathbf{x}_A) = 0.05$:
$$P(Y=1 \mid \mathbf{x}_E) = 1 - 8(0.05) = 0.60$$
$$0.25 P(Y=1 \mid \mathbf{x}_B) - 0.40(0.05) = -0.01 \implies 0.25 P(Y=1 \mid \mathbf{x}_B) = 0.01 \implies P(Y=1 \mid \mathbf{x}_B) = 0.04$$

# Q. Asset Pricing and Portfolio Optimization.



In standard finance, models often assume asset returns follow a perfect Normal (Gaussian) distribution. However, real-world markets experience extreme events, leading to "fat tails" and skewed returns. MaxEnt allows an analyst to infer the **least-biased probability distribution** of an asset's future price based *only* on what the market currently tells us, without forcing it into a rigid Gaussian shape.

**Background:**

You are a quantitative portfolio manager evaluating a highly volatile tech stock, **Asset X**, currently trading at $\$100$. You want to determine the risk-neutral probability distribution of its price, $S_T$, in exactly one month. This distribution will allow you to correctly price exotic derivatives and optimize your portfolio's risk exposure.

**Available Market Data (The Constraints):**

You do not know the true distribution, but you can observe the following macroeconomic aggregates and market-clearing prices from the options exchange:

1. **The Risk-Free Growth:** Based on the risk-free interest rate, the expected future price of the asset must match the forward price:

$$\mathbb{E}[S_T] = \$102$$


2. **The Market Volatility:** Based on standard historical variance, the expected squared price is constrained by:

$$\mathbb{E}[S_T^2] = 10,600$$


3. **Asymmetric Market Sentiment (The "Skew"):** Due to heavy trading on out-of-the-money put options, the market is pricing in a strong downside risk. The expected value of a specific logarithmic penalty (which captures downward crash risk) is constrained to:

$$\mathbb{E}[\ln(S_T)] = 4.60$$



**Objective:**

Formulate and solve the Maximum Entropy problem to find the continuous probability density function $p(S_T)$ for $S_T > 0$.



## ** Answers - Asset Pricing and Portfolio Optimization via Maximum Entropy**

To find the least-biased probability density function $p(S_T)$ for the future stock price $S_T > 0$, we maximize the differential entropy subject to the given market constraints using **Lagrange Multipliers**.

---

### **1. The Optimization Framework**

The objective is to maximize the Shannon differential entropy:
$$H(p) = -\int_{0}^{\infty} p(S_T) \ln p(S_T) \, dS_T$$

Subject to the following four constraints (including total probability normalization):
1. **Normalization:** $\int_{0}^{\infty} p(S_T) \, dS_T = 1$
2. **Mean (Forward Price):** $\int_{0}^{\infty} S_T p(S_T) \, dS_T = 102$
3. **Second Moment (Volatility):** $\int_{0}^{\infty} S_T^2 p(S_T) \, dS_T = 10,600$
4. **Market Sentiment (Skew):** $\int_{0}^{\infty} \ln(S_T) p(S_T) \, dS_T = 4.60$

---

### **2. Deriving the Functional Form**

We construct the Lagrangian function $L$ with multipliers $\lambda_0, \lambda_1, \lambda_2,$ and $\lambda_3$:

$$L(p) = -\int_{0}^{\infty} p(S_T) \ln p(S_T) \, dS_T - \lambda_0 \left(\int_{0}^{\infty} p(S_T) \, dS_T - 1\right) - \lambda_1 \left(\int_{0}^{\infty} S_T p(S_T) \, dS_T - 102\right) - \lambda_2 \left(\int_{0}^{\infty} S_T^2 p(S_T) \, dS_T - 10,600\right) - \lambda_3 \left(\int_{0}^{\infty} \ln(S_T) p(S_T) \, dS_T - 4.60\right)$$

Taking the functional derivative with respect to $p(S_T)$ and setting it to zero:

$$\frac{\partial L}{\partial p(S_T)} = - \ln p(S_T) - 1 - \lambda_0 - \lambda_1 S_T - \lambda_2 S_T^2 - \lambda_3 \ln(S_T) = 0$$

Solving for $p(S_T)$:

$$p(S_T) = \exp(-1 - \lambda_0) \cdot S_T^{-\lambda_3} \cdot \exp\left(-\lambda_1 S_T - \lambda_2 S_T^2\right)$$

Letting $C = \exp(-1 - \lambda_0)$ serve as the normalization constant, the explicit MaxEnt density function is:

$$p(S_T) = C \cdot S_T^{-\lambda_3} \exp\left(-\lambda_1 S_T - \lambda_2 S_T^2\right), \quad \text{for } S_T > 0$$

---

### **3. Economic Interpretation of the Resulting Distribution**

The inferred distribution is a highly flexible form of a **Generalized Gamma / Modified Log-Normal Distribution**. It naturally adapts away from a rigid Gaussian curve to account for the asymmetric risk:

* **$\exp(-\lambda_2 S_T^2)$**: Controls the rapid decay at ultra-high price levels, capping extreme upside risk based on bounded asset volatility.
* **$S_T^{-\lambda_3}$**: Directly incorporates the market skew. Because $\mathbb{E}[\ln(S_T)] = 4.60$ is lower than the log of the expected forward price ($\ln(102) \approx 4.625$), $\lambda_3$ acts to pull weight toward the lower tail, forming the **"volatility smile/skew"** observed empirically during market stress.

---

### **4. Numerical Implementation Approach**

Since the system of integrals cannot be solved analytically in closed form due to the simultaneous presence of $S_T$, $S_T^2$, and $\ln(S_T)$, the Lagrange multipliers must be found numerically by minimizing the convex dual partition function $\Gamma(\lambda_1, \lambda_2, \lambda_3)$:

$$\min_{\lambda_1, \lambda_2, \lambda_3} \left[ \ln \int_{0}^{\infty} S_T^{-\lambda_3} \exp\left(-\lambda_1 S_T - \lambda_2 S_T^2\right) dS_T + 102\lambda_1 + 10,600\lambda_2 + 4.60\lambda_3 \right]$$

This optimization is typically computed in quantitative workflows using Newton-Raphson algorithms via `scipy.optimize`.
